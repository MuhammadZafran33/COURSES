```python
import pandas as pd
import numpy as np
```

## 2. Load Dataset

```python
df = pd.read_csv("housing.csv")   
```

### 3. Check Data (Before Cleaning)

```python
print(df.head())     
print(df.info())     
print(df.describe()) 
```

### 4. Data Cleaning Steps
####  (1) Handle Missing Values

```python

print(df.isnull().sum())


df.fillna(df.mean(numeric_only=True), inplace=True)
```

#### (2) Remove Duplicates

```python
df.drop_duplicates(inplace=True)
```

#### (3) Fix Data Types

```python

df["housing_median_age"] = df["housing_median_age"].astype(int)

df["population"] = pd.to_datetime(df["population"])
```

#### (4) Fix Column Names

```python

df.columns = df.columns.str.lower()

df.columns = df.columns.str.replace(" ", "_")
```

#### (6) Handle Outliers (Simple Method)

```python

df = df[df["housing_median_age"] < 50]
```

### Summary

```python
print(df.info())
print(df.describe())
```

### 6. Save Cleaned Data

```python
df.to_csv("cleaned_data.csv", index=False)
```
