# Python Program: Spam Email Detection

```python
# open the file and read all emails
file = open("email.txt", "r")
emails = file.readlines()
file.close()

# keywords to detect spam emails
spam_keywords = ["win", "free", "prize", "money", "claim"]

all_emails = []
labels = []
spam_emails = []

# process each email
for email in emails:
    email = email.strip()
    all_emails.append(email)

    email_lower = email.lower()

    is_spam = False

    # check if any spam word exists in email
    for word in spam_keywords:
        if word in email_lower:
            is_spam = True

    # assign label
    if is_spam:
        labels.append("Spam")
        spam_emails.append(email)
    else:
        labels.append("Not Spam")

# display all emails
print("All Emails:")
for e in all_emails:
    print(e)

# display labels
print("\nEmail Labels:")
for i in range(len(all_emails)):
    print(all_emails[i], "->", labels[i])

# display only spam emails
print("\nSpam Emails:")
for s in spam_emails:
    print(s)
```
