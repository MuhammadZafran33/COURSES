# PART 1 — What is Matplotlib?
---

## The Story

In **2003**, a scientist named **John Hunter** was working at a hospital studying brain signals (EEG data) from epilepsy patients. He needed to draw charts to visualise this data.

He was using a paid software called **MATLAB** — but the hospital could not afford licences for everyone. So he did something remarkable — he built his own charting tool using Python and made it **completely free**.

He called it **Matplotlib** — MATLAB + plot + lib (library).

Today, Matplotlib is downloaded **millions of times every month** and is the foundation of almost all Python visualisation.

---

## What is Matplotlib?

Matplotlib is a Python library for creating **charts, graphs, and plots**.

Think of it like **Microsoft Paint for data** — you have full control over everything you draw. Every color, every label, every line — you control it manually.

---

## Why Do We Need Matplotlib?

Raw data is just numbers. Humans are terrible at understanding hundreds of numbers by reading them. But we are excellent at understanding **pictures and patterns**.

```
Without chart — just numbers:          With a chart:
56, 72, 91, 43, 85, 78, 66...          You instantly see the distribution!
34, 88, 71, 55, 94, 47, 83...          High scorers, low scorers, average...
... (1000 numbers)                      All visible in 2 seconds.
```

Matplotlib turns numbers into pictures.

---

## Real World Use Cases

- **Data Scientists** — explore datasets before building ML models
- **Researchers** — create charts for scientific papers
- **Machine Learning** — plot training loss and accuracy over time

- **Students** — understand data visually

---

## In Artificial Intelligence

- Plot how model accuracy improves during training (learning curve)
- Visualise which features matter most for predictions
- Compare performance of different models side by side
- Show the confusion matrix results visually

### Chart 1 — Histogram

**What it is:** A histogram shows how values are distributed — where are most of them? Are they spread out or bunched together?

**When to use:** When you have one numeric column and want to see its distribution.

**Real world use:** How are exam scores distributed? Are most students getting 60-80 or are scores spread everywhere?

**Data needed:** One numeric column

````python
# Histogram — Exam Score Distribution
# bins = how many bars (buckets) to divide the data into

plt.figure(figsize=(8, 5))
plt.hist(df['exam_score'], bins=20)
plt.title('Exam Score Distribution')
plt.xlabel('Exam Score')
plt.ylabel('Number of Students')
plt.show()
````

### What We See
Most students score between 55 and 85. The tallest bar shows the most common score range. This is a roughly bell-shaped distribution — typical for exam scores.

**Try it yourself:** Change `bins=20` to `bins=10` or `bins=40` and see how the shape changes.

````python
# Histogram — Study Hours per Day

plt.figure(figsize=(8, 5))
plt.hist(df['study_hours_per_day'], bins=15)
plt.title('Study Hours per Day Distribution')
plt.xlabel('Hours Studied per Day')
plt.ylabel('Number of Students')
plt.show()
````

### Chart 2 — Bar Chart

**What it is:** A bar chart shows how many items are in each category, or the value for each category.

**When to use:** When you have categories (like gender, diet quality) and want to compare them.

**Real world use:** How many students are male vs female? Which diet quality is most common?

**Data needed:** One categorical column

````python
gender = df['gender'].value_counts()
gender
````

````python
# Bar Chart — Number of students by gender
# First we count how many students are in each gender group

gender_counts = df['gender'].value_counts()

plt.figure(figsize=(7, 5))
plt.bar(gender_counts.index, gender_counts.values)
plt.title('Number of Students by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Students')
plt.show()
````

````python
avg_score = df.groupby('gender')['exam_score'].mean()
print(avg_score)
````

````python
# Bar Chart — Average exam score by gender
# groupby calculates the average for each gender group

avg_score = df.groupby('gender')['exam_score'].mean()

plt.figure(figsize=(7, 5))
plt.bar(avg_score.index, avg_score.values)
plt.title('Average Exam Score by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Exam Score')
plt.show()
````

### What We See
All three gender groups score very similarly on average. Gender alone does not determine exam performance.

### Chart 3 — Scatter Plot

**What it is:** A scatter plot puts one variable on x-axis and another on y-axis. Each dot is one student. The pattern of dots shows if the two variables are related.

**When to use:** When you have TWO numeric columns and want to see if they are related.

**Real world use:** Does studying more lead to higher exam scores?

**Data needed:** Two numeric columns

**Patterns:**
- Dots going bottom-left to top-right → positive relationship
- Dots going top-left to bottom-right → negative relationship
- Random dots with no pattern → no relationship

````python
# Scatter Plot — Study Hours vs Exam Score
# Each dot = one student

plt.figure(figsize=(8, 5))
plt.scatter(df['study_hours_per_day'], df['exam_score'])
plt.title('Study Hours vs Exam Score')
plt.xlabel('Study Hours per Day')
plt.ylabel('Exam Score')
plt.show()
````

### What We See
Clear upward trend — dots go from bottom-left to top-right. Students who study more score higher. This is a strong positive relationship (correlation = 0.83).

````python
# Scatter Plot — Social Media Hours vs Exam Score

plt.figure(figsize=(8, 5))
plt.scatter(df['social_media_hours'], df['exam_score'])
plt.title('Social Media Hours vs Exam Score')
plt.xlabel('Social Media Hours per Day')
plt.ylabel('Exam Score')
plt.show()
````

### What We See
Slight downward trend — more social media hours is associated with slightly lower scores. The relationship is weaker than study hours though.

### Chart 5 — Pie Chart

**What it is:** A pie chart shows proportions — how a total is divided into parts.

**When to use:** When you have 2 to 5 categories and want to show each one's percentage of the whole.

**Real world use:** What percentage of students have a part-time job vs not?

**Data needed:** Categorical column with 2 to 5 categories

**Important:** Do NOT use pie charts with many categories — slices become too small to read. Use a bar chart instead.

### Chart 7 — Box Plot

**What it is:** A box plot shows the spread of values and highlights outliers. It shows 5 things at once: minimum, Q1, median, Q3, maximum.

**When to use:** When you want to understand the spread of a numeric column and spot outliers.

**Reading a box plot:**
```
  Line at top of box  = Q3 (75% of data is below this)
  Line inside box     = Median (50% below, 50% above)
  Line at bottom      = Q1 (25% of data is below this)
  Whiskers            = extend to min and max (within limits)
  Dots beyond whiskers = outliers
```

**Data needed:** One or more numeric columns

````python
# Box Plot — Exam Score

plt.figure(figsize=(6, 7))
plt.boxplot(df['exam_score'])
plt.title('Exam Score — Box Plot')
plt.ylabel('Exam Score')
plt.show()
````

## The Limitation of Matplotlib

Matplotlib is powerful but has one big weakness — **comparing groups is hard**.

For example — what if you want to draw a histogram for male students AND female students on the same chart to compare them? In Matplotlib, you would need to:
1. Filter male students manually
2. Filter female students manually
3. Draw both on the same axes
4. Add a legend manually

That is 4 steps. With Seaborn — it is ONE line.

This is where Seaborn comes in.

# PART 4 — What is Seaborn?
---

## The Story

By **2012**, Matplotlib existed and worked well. But data scientists had a daily complaint:

*"Why does it take 20 lines of code just to compare exam scores between male and female students? I just want to see the difference quickly!"*

**Michael Waskom**, a researcher at Stanford University, solved this problem. He built **Seaborn** — a library that sits on top of Matplotlib and makes statistical charts easy and beautiful by default.

He released it in **2012** and named it `sns` (short for Samuel Norman Seaborn — a character from the TV show The West Wing).

---

## What is Seaborn?

Seaborn is a Python visualisation library **built on top of Matplotlib**. It does not replace Matplotlib — it uses Matplotlib underneath but hides the complicated setup from you.

Think of it like this:
- **Matplotlib** = Microsoft Paint — full control, you do everything manually
- **Seaborn** = Canva — beautiful results with minimal effort

---

## Why Do We Need Seaborn?

Three reasons:

**1. Less code for the same result**
```python
# Matplotlib — compare exam scores by gender (manual work)
male   = df[df['gender']=='Male']['exam_score']
female = df[df['gender']=='Female']['exam_score']
other  = df[df['gender']=='Other']['exam_score']
plt.hist(male,   alpha=0.5, label='Male')
plt.hist(female, alpha=0.5, label='Female')
plt.hist(other,  alpha=0.5, label='Other')
plt.legend()
plt.show()   # 7 lines!

# Seaborn — same result in 1 line
sns.histplot(data=df, x='exam_score', hue='gender')
```

**2. Beautiful default style**  
Seaborn charts look professional without any extra styling code.

**3. Statistical charts built-in**  
Charts like violin plot, pair plot, heatmap, regression plot — Seaborn has these ready. Matplotlib does not.

---

## The hue= Parameter — Seaborn's Superpower

The most important Seaborn feature is the `hue=` parameter.

It lets you split any chart by a category — automatically coloring each group differently.

```python
# Without hue — one color for everyone
sns.histplot(data=df, x='exam_score')

# With hue — different color per gender, automatically!
sns.histplot(data=df, x='exam_score', hue='gender')
```

This is what Matplotlib cannot do in one line.

# PART 5 — What Can Seaborn Do That Matplotlib Cannot?
---

| Feature | Matplotlib | Seaborn |

### Seaborn Chart 1 — histplot (Histogram)

Same as `plt.hist()` but:
- Works directly with DataFrame (`data=df`)
- Can add a KDE smooth curve with `kde=True`
- Can split by group with `hue=`

````python
sns.histplot(data=df, x='exam_score', kde=True)
````

````python
# Seaborn Histogram — Exam Score
# kde=True adds a smooth curve showing the shape

plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='exam_score', kde=True)
plt.title('Exam Score Distribution')
plt.xlabel('Exam Score')
plt.ylabel('Count')
plt.show()
````

````python
# Seaborn Histogram — Split by gender using hue
# This is what Matplotlib cannot do in one line!

plt.figure(figsize=(9, 5))

sns.histplot(data=df, x='exam_score', hue='gender', kde=True)

plt.title('Exam Score Distribution by Gender')
plt.xlabel('Exam Score')
plt.ylabel('Count')
plt.show()
````

### What We See
Three overlapping distributions — one per gender. They all follow a very similar shape, confirming that gender does not strongly affect exam performance. Seaborn coloured each group automatically and added a legend — in ONE line of code.

### Seaborn Chart 2 — countplot

**What it is:** countplot automatically counts how many students are in each category — you do not need to calculate counts yourself.

**Compare with Matplotlib bar chart:**
```python
# Matplotlib — you count manually first
counts = df['diet_quality'].value_counts()
plt.bar(counts.index, counts.values)

sns.countplot(data=df, x='diet_quality')
```

````python
# Count Plot with hue — Diet Quality split by Part-Time Job
# Two categories compared in one chart

plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='diet_quality', hue='part_time_job', order=['Poor', 'Fair', 'Good'])
plt.title('Diet Quality by Part-Time Job Status')
plt.xlabel('Diet Quality')
plt.ylabel('Count')
plt.show()
````

### Seaborn Chart 3 — barplot

**What it is:** Shows the average value of a numeric column for each category. Seaborn calculates the mean automatically AND adds error bars showing uncertainty.

**When to use:** When you want to compare averages across groups.

````python
# Bar Plot — Average exam score by diet quality
# Seaborn calculates the mean for each group automatically
# The small lines on top of bars are error bars (uncertainty)

plt.figure(figsize=(7, 5))
sns.barplot(data=df, x='diet_quality', y='exam_score', order=['Poor', 'Fair', 'Good'])
plt.title('Average Exam Score by Diet Quality')
plt.xlabel('Diet Quality')
plt.ylabel('Average Exam Score')
plt.show()
````

````python
# Bar Plot — Average exam score by internet quality

plt.figure(figsize=(7, 5))
sns.barplot(data=df, x='internet_quality', y='exam_score', order=['Poor', 'Average', 'Good'])
plt.title('Average Exam Score by Internet Quality')
plt.xlabel('Internet Quality')
plt.ylabel('Average Exam Score')
plt.show()
````

### Seaborn Chart 4 — scatterplot and regplot

**scatterplot** — same as plt.scatter() but with `hue=` support

**regplot** — scatter plot PLUS a trend line (regression line) automatically drawn

````python
# Seaborn Scatter Plot — Study Hours vs Exam Score
# alpha makes dots transparent so overlapping dots are visible

plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='study_hours_per_day', y='exam_score', alpha=0.4)
plt.title('Study Hours vs Exam Score')
plt.xlabel('Study Hours per Day')
plt.ylabel('Exam Score')
plt.show()
````

````python
# regplot — Scatter with Trend Line
# The line shows the overall direction of the relationship
# This is exactly what Linear Regression does!

plt.figure(figsize=(8, 5))
sns.regplot(data=df, x='study_hours_per_day', y='exam_score', scatter_kws={'alpha':0.3})
plt.title('Study Hours vs Exam Score (with Trend Line)')
plt.xlabel('Study Hours per Day')
plt.ylabel('Exam Score')
plt.show()
````

### What We See
The blue line going upward is the trend line. The shaded area around it is the confidence band — how certain we are about the trend. The steeper the line, the stronger the relationship.

This trend line is exactly what Linear Regression learns!

````python
# Scatter with hue — 3 variables in one chart!
# x-axis = study hours, y-axis = exam score, color = diet quality

plt.figure(figsize=(9, 6))
sns.scatterplot(data=df, x='study_hours_per_day', y='exam_score',
                hue='diet_quality', alpha=0.5)
plt.title('Study Hours vs Exam Score — Coloured by Diet Quality')
plt.xlabel('Study Hours per Day')
plt.ylabel('Exam Score')
plt.show()
````

### Seaborn Chart 6 — violinplot

**What it is:** A violin plot combines a box plot and a histogram in one chart. The width at each height shows how many students have that score — wider = more students.

**This does not exist in Matplotlib** — it is a Seaborn-only chart.

**When to use:** When you want to see the full shape of the distribution for each group, not just the box statistics.

````python
# Violin Plot — Exam Score by Gender
# Wide part = more students at that score
# Narrow part = fewer students at that score

plt.figure(figsize=(8, 6))
sns.violinplot(data=df, x='gender', y='exam_score')
plt.title('Exam Score Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Exam Score')
plt.show()
````

````python
# Violin Plot — Exam Score by Internet Quality

plt.figure(figsize=(8, 6))
sns.violinplot(data=df, x='internet_quality', y='exam_score', order=['Poor', 'Average', 'Good'])
plt.title('Exam Score Distribution by Internet Quality')
plt.xlabel('Internet Quality')
plt.ylabel('Exam Score')
plt.show()
````

### Seaborn Chart 7 — heatmap

**What it is:** A heatmap shows a grid of values using colours. Dark green = strong positive relationship. Dark red = strong negative relationship. Yellow = no relationship.

**Most common use:** Showing the correlation matrix — how every numeric column relates to every other numeric column.

**This does not exist in Matplotlib** — it is a Seaborn-only chart.

**Critical for Machine Learning:** Before training any model, always look at the heatmap to find which features are most related to your target variable.

````python
# Heatmap — Correlation Matrix
# corr() calculates correlation between all numeric columns
# annot=True shows the numbers inside each cell

numeric_df = df.select_dtypes(include='number')
corr_matrix = numeric_df.corr()

plt.figure(figsize=(10, 7))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='RdYlGn', center=0)
plt.title('Correlation Between All Numeric Columns')
plt.show()
````

### How to Read This
- **Look at the exam_score row** — which cell is the most green?
- `study_hours_per_day` has the strongest green (0.83) — the strongest positive relationship with exam score
- `social_media_hours` and `netflix_hours` are slightly red — more screen time = slightly lower scores
- The diagonal is always 1.0 — a column perfectly correlates with itself

**AI Use:** The greenest cells in the exam_score row tell you which features will be most useful in your ML model.

### Seaborn Chart 8 — pairplot

**What it is:** pairplot draws scatter plots for every combination of numeric columns. It gives you a complete overview of all relationships at once.

**This does not exist in Matplotlib** — it is a Seaborn-only chart.

**When to use:** Early in data exploration when you want to quickly see all relationships.

````python
# Pair Plot — Select a few columns to keep it readable
# Diagonal shows histogram for each column
# Other cells show scatter plots between two columns

selected_cols = ['study_hours_per_day', 'sleep_hours', 'social_media_hours', 'exam_score']

sns.pairplot(df[selected_cols])
plt.suptitle('Pair Plot — All Relationships at Once', y=1.02)
plt.show()
````

### What We See
- **Diagonal** = histogram for each column (its own distribution)
- **Other cells** = scatter plot between two columns
- The `study_hours_per_day` vs `exam_score` cell clearly shows an upward trend
- Most other pairs show scattered dots with no clear pattern

---
### Seaborn Chart 9 — lineplot

Seaborn's line plot automatically calculates the mean and shows a confidence band. With `hue=` it draws separate lines per group.

````python
# Line Plot — Average exam score by age
# Seaborn calculates the mean automatically for each age
# The shaded area shows the confidence interval

plt.figure(figsize=(9, 5))
sns.lineplot(data=df, x='age', y='exam_score')
plt.title('Average Exam Score by Age')
plt.xlabel('Age')
plt.ylabel('Exam Score')
plt.show()
````

````python
# Line Plot — Average study hours by age, split by gender

plt.figure(figsize=(9, 5))
sns.lineplot(data=df, x='age', y='study_hours_per_day', hue='gender')
plt.title('Average Study Hours by Age and Gender')
plt.xlabel('Age')
plt.ylabel('Study Hours per Day')
plt.show()
````

# Final Summary — Matplotlib vs Seaborn
---

## When to Use Which

| Task | Use Matplotlib | Use Seaborn |
|------|---------------|-------------|
| Simple single chart | ✅ | ✅ |
| Compare groups with one line | ❌ Manual | ✅ `hue=` |
| Pie chart | ✅ `plt.pie()` | ❌ Not available |
| Violin plot | ❌ Not available | ✅ `sns.violinplot()` |
| Heatmap | ❌ Not available | ✅ `sns.heatmap()` |
| Pair plot | ❌ Not available | ✅ `sns.pairplot()` |
| Regression line | ❌ Manual math | ✅ `sns.regplot()` |
| Auto counting | ❌ Calculate first | ✅ `sns.countplot()` |
| Full custom control | ✅ | ❌ Limited |

## The Golden Rule

> **Use Seaborn first.** If Seaborn cannot do what you need — then use Matplotlib.  
> In practice, most data scientists use both in the same notebook.

## Chart Selection Guide

| Question | Chart | Library |
|----------|-------|--------|
| How are values distributed? | Histogram | Both |
| How many in each group? | Bar / Count Plot | Both |
| What is each group's average? | Bar Plot | Seaborn |
| Are two things related? | Scatter Plot | Both |
| What is the trend? | Line Chart / regplot | Both |
| What is each group's share? | Pie Chart | Matplotlib |
| What is the spread and outliers? | Box Plot | Both |
| What is the full distribution shape? | Violin Plot | Seaborn |
| How do all columns relate? | Heatmap | Seaborn |
| All vs all in one go? | Pair Plot | Seaborn |
