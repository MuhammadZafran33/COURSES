## 📜 What is Pandas — and Where Did It Come From?

In **2008**, a programmer named **Wes McKinney** was working at a finance company in New York. Every day he had to analyse huge tables of data — stock prices, economic numbers, reports.

The tools available at the time were painful to use for this kind of work. So Wes built his own — and called it **pandas**.

The name comes from **"panel data"**, an economics term for data tracked over time. It also stands for **"Python Data Analysis"**.

He released it for free in **2009**. Today it is downloaded **100 million times every month** and used by Google, Netflix, NASA, and almost every data team in the world.

# 📊 Introduction to Pandas

## 📌 What is Pandas?
Pandas is a powerful Python library used for **data analysis and data manipulation**.  
It provides easy-to-use data structures for handling structured data like tables.

---

## 📌 Key Data Structures

- **Series** → 1D labeled array (like a column)  
- **DataFrame** → 2D table (rows and columns, like Excel sheet)

---

## 📌 Key Features of Pandas

- **Easy data handling** → Load, clean, and process data efficiently  
- **Data alignment** → Automatically aligns data by labels  
- **Handling missing data** → Provides functions to detect and fill missing values  
- **Filtering & selection** → Easily select specific rows and columns  
- **Grouping & aggregation** → Perform operations like sum, mean, count  
- **Merging & joining** → Combine multiple datasets  
- **Time series support** → Work with date and time data  
- **File handling** → Read/write data from CSV, Excel, JSON, HTML, SQL etc.

---

## 📌 Why Use Pandas?

- Simplifies complex data operations  
- Saves time compared to manual coding  
- Widely used in Data Science and AI  

---

## 📌 Applications in AI

- Data preprocessing  
- Cleaning datasets  
- Feature selection  
- Preparing data for machine learning models  

---

## 🤔 Why Do We Need Pandas?

Imagine you have a spreadsheet with 1,000 student records. You want to:
- Find all students who scored above 80
- Calculate the average score by gender
- Fix typos in the data

In plain Python you would write 30+ lines of code and loops.

With Pandas, each of those tasks is **one line**.

---

## 🏗️ The Two Main Building Blocks

**Series** → a single column of data (like one column in Excel)
```
0    85
1    72
2    91
3    68
```

**DataFrame** → a full table with rows and columns (like an entire Excel sheet)
```
   name   age  score
0  Alice   21     85
1  Bob     19     72
2  Carol   22     91
```

Everything we do in Pandas works with these two objects.

# Part 4 — Cleaning the Data

Real data is almost always messy. Before we analyse it, we must fix it.

Here are the problems we found in our dataset:

| Problem | Example |
|---------|--------|
| Column names have spaces | `'student_id '` instead of `'student_id'` |
| Inconsistent casing | Some columns are UPPERCASE, others lowercase |
| Duplicate rows | Same student appears twice |
| Useless column | `RANDOM_ID` has no meaning |
| Missing values | Many columns have NaN (blank) cells |
| Gender typos | `'femle'`, `'mle'`, `' FEMALE '` all mean different things to pandas |
| Text instead of numbers | `'ten'` written as a word in a numeric column |
| Negative hours | `-3.5` hours of Netflix is impossible |
| Exam scores > 100 | A score of 365 is impossible |

We will fix each one step by step.

---

### Step 22 — Fix Column Names (Remove Spaces)

Some column names have hidden spaces at the start or end. This causes bugs like `df['student_id']` giving a KeyError even though you can see the column.

`str.strip()` removes spaces from both ends.  
`str.lower()` makes everything lowercase — so all column names are consistent.

````python
# Remove spaces and make all column names lowercase
df.columns = df.columns.str.strip().str.lower()
````

````python
# Verify — all clean now
df.columns
````

````python
df.head()
````

**Fix Text Numbers in Numeric Columns**

The `study_hours_per_day` and `sleep_hours` columns are supposed to be numbers, but some entries have the word `'ten'` instead of `10`.

`replace()` swaps the text for the correct number string.  
`pd.to_numeric()` then converts the whole column to a proper number type.  
`errors='coerce'` means: if anything still can't be converted, make it NaN instead of crashing.

````python
# # Replace the word 'ten' with the number '10'

df['study_hours_per_day'] = df['study_hours_per_day'].replace('ten', '10')


df['sleep_hours']         = df['sleep_hours'].replace('ten', '10')
````

````python
# Convert to numeric (number) type

df['study_hours_per_day'] = pd.to_numeric(df['study_hours_per_day'], errors='coerce')
df['sleep_hours']         = pd.to_numeric(df['sleep_hours'],         errors='coerce')
````

````python
df['sleep_hours'].value_counts()
````

````python
df.info()
````

### Step 23 — Drop a Useless Column

The `random_id` column is just random numbers with no meaning. It has nothing to do with student performance. We remove it with `drop()`.

`columns=['random_id']` tells pandas we are dropping a column (not a row).

````python
# Remove the random_id column

df = df.drop(columns=['random_id'])
````

````python
# Verify — column is gone, shape should now be 16 columns
df.shape
````

````python
df.columns
````

### Step 24 — Remove Duplicate Rows

`drop_duplicates()` removes rows that are exact copies of another row.  
`keep='first'` keeps the first occurrence and removes the rest.  
`reset_index(drop=True)` renumbers the rows cleanly after removal.

````python
# Remove duplicate rows

df = df.drop_duplicates(keep='first')
df = df.reset_index(drop=True)
````

````python
# Verify — should be 0 now
df.duplicated().sum()
````

### Step 25 — Drop Rows with Missing Student ID

If a student has no ID, we cannot identify them. These rows are useless to us.

`dropna(subset=['student_id'])` drops only the rows where `student_id` is missing (NaN).

````python
df= df.drop(columns=['student_id'])
````

### Step 26 — Drop Rows with Missing Exam Score

The exam score is the most important column — it's what we want to understand. Rows with no exam score are not useful for our analysis.

````python
# Drop rows where exam_score is missing

df = df.dropna(subset=['exam_score'])
df = df.reset_index(drop=True)
````

````python
# How many rows do we have now?
df.shape
````

### Step 27 — Fill Missing Numbers with the Median

For numeric columns, we replace missing values with the **median** (the middle value).  
We use median instead of mean because our data has outliers — extreme values can pull the mean far from the typical value, but the median stays representative.

`fillna()` replaces NaN with whatever value you give it.

````python
# Fill missing numeric values with the median of each column
df['age']                  = df['age'].fillna(df['age'].median())
df['social_media_hours']   = df['social_media_hours'].fillna(df['social_media_hours'].median())
df['netflix_hours']        = df['netflix_hours'].fillna(df['netflix_hours'].median())
df['attendance_percentage']= df['attendance_percentage'].fillna(df['attendance_percentage'].median())
df['exercise_frequency']   = df['exercise_frequency'].fillna(df['exercise_frequency'].median())
df['mental_health_rating'] = df['mental_health_rating'].fillna(df['mental_health_rating'].median())
````

````python
df.isnull().sum().sum()
````

````python
# Check — these columns should now have 0 missing values
df[['age', 'social_media_hours', 'netflix_hours']].isnull().sum()
````

### Step 28 — Fill Missing Text with the Mode

For text (categorical) columns, we use the **mode** — the most frequently occurring value.  
`mode()[0]` gets the most common value (mode returns a Series, so `[0]` grabs the first/top item).

````python
df['part_time_job']               = df['part_time_job'].fillna(df['part_time_job'].mode())
````

````python
df.isnull().sum()
````

````python
# Fill missing text columns with the most common value
# [0] is used to get the first (most frequent) value from mode() result.

df['part_time_job']               = df['part_time_job'].fillna(df['part_time_job'].mode()[0])
df['diet_quality']                = df['diet_quality'].fillna(df['diet_quality'].mode()[0])
df['internet_quality']            = df['internet_quality'].fillna(df['internet_quality'].mode()[0])
df['extracurricular_participation']= df['extracurricular_participation'].fillna(df['extracurricular_participation'].mode()[0])
````

````python
#We cannot guess personal identity attributes —
# filling with the mode would fabricate demographic data, introducing bias into the AI model.
````

````python
# Missing categorical values are filled with "Unknown" to avoid wrong assumptions and keep missing data information...Added Bias.
# Fill gender and parental education with 'Unknown'

df['gender']                   = df['gender'].fillna('Unknown')
df['parental_education_level'] = df['parental_education_level'].fillna('Unknown')
````

### Step 29 — Fix Spaces in the Gender Column

The gender column is very messy. It has:
- Leading/trailing spaces: `' male '` instead of `'male'`
- Mixed case: `'MALE'`, `'Male'`, `'male'`
- Typos: `'mle'` instead of `'male'`, `'femle'` instead of `'female'`

We first strip spaces and make everything lowercase to reduce the chaos.

````python
# Strip spaces and make lowercase
df['gender'] = df['gender'].str.strip().str.lower()
````

````python
# See what values remain after stripping and lowercasing
df['gender'].value_counts()
````

### Step 30 — Map Gender to Clean Standard Values

Now we use a **dictionary** to map every messy variant to the correct clean value.  
`map()` replaces each value using the dictionary — like a lookup table.

````python
# Map all messy variants to clean standard values
gender_map = {
    'female'  : 'Female',
    'femle'   : 'Female',
    'f'       : 'Female',
    'male'    : 'Male',
    'mle'     : 'Male',
    'm'       : 'Male',
    'other'   : 'Other',
    'unknown' : 'Unknown'
}

df['gender'] = df['gender'].map(gender_map).fillna('Unknown')
````

````python
# Verify — should now have only 4 clean values
df['gender'].value_counts()
````

### Step 31 — Fix Spaces in Other Categorical Columns

Other text columns also have leading/trailing spaces causing the same problem — `'No'` and `' No'` look different to pandas but mean the same thing.

`str.strip()` removes the spaces. `str.title()` makes it Title Case (first letter uppercase).

````python
# Strip spaces and standardise casing for all these columns
text_columns = [
    'part_time_job',
    'diet_quality',
    'internet_quality',
    'extracurricular_participation',
    'parental_education_level'
]

for col in text_columns:
    df[col] = df[col].str.strip().str.title()
````

````python
# Verify — should show clean values like 'Yes', 'No' not ' Yes' and ' No'
df['part_time_job'].value_counts()
````

### Step 32 — Fix Text Numbers in Numeric Columns

The `study_hours_per_day` and `sleep_hours` columns are supposed to be numbers, but some entries have the word `'ten'` instead of `10`.

`replace()` swaps the text for the correct number string.  
`pd.to_numeric()` then converts the whole column to a proper number type.  
`errors='coerce'` means: if anything still can't be converted, make it NaN instead of crashing.

````python
# Fill any remaining NaN created during conversion
df['study_hours_per_day'] = df['study_hours_per_day'].fillna(df['study_hours_per_day'].median())
df['sleep_hours']         = df['sleep_hours'].fillna(df['sleep_hours'].median())
````

````python
# Verify data type is now float64, not object
df[['study_hours_per_day', 'sleep_hours']].dtypes
````

### Step 33 — Remove Negative Hours (Impossible Values)

You cannot spend a negative number of hours watching Netflix or on social media. These are data entry errors.

`clip(lower=0)` replaces any value below 0 with 0. It **clips** the values to a valid range.

````python
# Set the minimum value to 0 — hours cannot be negative
df['netflix_hours']      = df['netflix_hours'].clip(lower=0)
df['social_media_hours'] = df['social_media_hours'].clip(lower=0)
df['exercise_frequency'] = df['exercise_frequency'].clip(lower=0)
df['sleep_hours']        = df['sleep_hours'].clip(lower=0, upper=16)
````

````python
# Verify — no more negative values in netflix_hours
df['netflix_hours'].min()
````

### Step 34 — Fix Exam Scores Above 100

An exam score cannot be above 100%. Some values are clearly wrong (e.g., 365, 502). We cap all scores at 100 using `clip()`.

````python
# Before — how many scores are above 100?
(df['exam_score'] > 100).sum()
````

````python
# Cap exam score between 0 and 100
df['exam_score'] = df['exam_score'].clip(lower=0, upper=100)
````

````python
# Verify — max should now be 100
df['exam_score'].max()
````

### Step 35 — Remove Rows with Unrealistic Age

Some students have an age of less than 15 — unrealistic for university students. We remove these rows entirely using boolean filtering.

````python
# Keep only rows where age is 15 or older
df = df[df['age'] >= 15]
df = df.reset_index(drop=True)
````

````python
# Verify — minimum age should now be 15 or above
df['age'].min()
````

### Step 36 — Check the Clean Dataset

Let's do a final check to see how much our data improved.

````python
# Final shape of the clean data
df.shape
````

````python
# Final missing value check — should be mostly 0
df.isnull().sum()
````

````python
# Look at the clean data
df.head()
````

### Step 39 — Filter Rows with a Condition

This is called **boolean masking**. We write a condition, and pandas returns only the rows where that condition is True.

Think of it like asking a question: "Which rows have exam_score above 85?"

````python
# Show only students who scored above 85
df[df['exam_score'] > 85]
````

### Step 40 — Filter with Two Conditions

Use `&` for AND (both must be true), `|` for OR (either must be true).  
⚠️ Always put each condition in its own parentheses `()`!

````python
# Students who scored above 80 AND study more than 5 hours
df[(df['exam_score'] > 80) & (df['study_hours_per_day'] > 5)]
````

### Step 41 — Filter with isin()

`isin()` filters rows where a value is in a given list. Like saying "give me rows where internet_quality is either Good or Average".

````python
# Students with Good or Average internet quality
df[df['internet_quality'].isin(['Good', 'Average'])]
````

### Step 42 — Create a Column with Simple Math

We can create a new column by doing arithmetic on existing columns. Pandas applies the operation to every row automatically.

````python
# Total screen time = social media + netflix
df['total_screen_time'] = df['social_media_hours'] + df['netflix_hours']
````

````python
# Check the new column
df[['social_media_hours', 'netflix_hours', 'total_screen_time']].head()
````

### Step 43 — Create a Column Using a Function

`apply()` runs a function on every row. We define a function that converts a score into a letter grade, then apply it to the entire column.

````python
# Define a function that converts score to grade
def get_grade(score):
    if score >= 85:   return 'A'
    elif score >= 70: return 'B'
    elif score >= 55: return 'C'
    elif score >= 40: return 'D'
    else:             return 'F'

# Apply it to every row in the exam_score column
df['grade'] = df['exam_score'].apply(get_grade)
````

````python
# See the grade distribution
df['grade'].value_counts()
````

### Step 44 — Create a Column Using pd.cut()

`pd.cut()` groups a continuous number into categories (bins).  
Here we split students into age groups.

````python
# Create age groups
df['age_group'] = pd.cut(
    df['age'],
    bins   = [14, 18, 21, 24, 35],
    labels = ['15-18', '19-21', '22-24', '25+']
)
````

````python
# See age group distribution
df['age_group'].value_counts().sort_index()
````

# Part 8 — Grouping and Aggregating

GroupBy is one of the most powerful tools in Pandas. It answers questions like:
- "What is the average score for each gender?"
- "How many students are in each grade?"

It works in 3 steps: **Split** → **Apply** → **Combine**

---

### Step 47 — Average Score by Gender

`groupby('gender')` splits the data into groups — one for each gender.  
Then we pick a column and apply `.mean()` to get the average within each group.

````python
# Average exam score for each gender
df.groupby('gender')['exam_score'].mean().round(2)
````

### Step 48 — Count Students in Each Grade

`count()` counts how many students fall in each group.

````python
# How many students got each grade?
df.groupby('grade')['student_id'].count()
````

### Step 49 — Multiple Statistics at Once with agg()

`agg()` lets you calculate several statistics for each group at the same time.  
This is great for building a summary table.

````python
# Summary table: count, avg score, avg study hours — grouped by grade
df.groupby('grade').agg(
    Students        = ('student_id', 'count'),
    Avg_Score       = ('exam_score', 'mean'),
    Avg_Study_Hours = ('study_hours_per_day', 'mean'),
    Avg_Sleep       = ('sleep_hours', 'mean')
).round(2)
````

### Step 50 — Pivot Table

A pivot table is like a cross-tabulation — rows are one category, columns are another, and cells are the value you want to see.

````python
# Average exam score by diet quality (rows) and internet quality (columns)
pd.pivot_table(
    df,
    values  = 'exam_score',
    index   = 'diet_quality',
    columns = 'internet_quality',
    aggfunc = 'mean'
).round(1)
````

### Step 51 — Correlation Between Columns

`corr()` measures how strongly two columns are related.
- Values close to **1.0** = strong positive relationship (one goes up, other goes up)
- Values close to **-1.0** = strong negative relationship (one goes up, other goes down)
- Values close to **0** = no clear relationship

````python
# Correlation of key columns with exam_score
cols = ['exam_score', 'study_hours_per_day', 'sleep_hours',
        'total_screen_time', 'attendance_percentage', 'mental_health_rating']

df[cols].corr()[['exam_score']].round(3)
````

### Step 52 — Cross Tabulation

`pd.crosstab()` counts how many students fall into each combination of two categories. Great for seeing patterns.

````python
# How many students of each gender got each grade?
pd.crosstab(df['gender'], df['grade'])
````
