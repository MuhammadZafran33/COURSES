## calculator Program

```python
operator = input("Enter an oprator (+ - * /)")
num1 = float(input("Enter the 1st number: "))
num2 = float(input("Enter the 2nd number: "))
if operator == "+":
    result = num1+num2
    print(round(result, 3))
elif operator == "-":
    result = num1-num2
    print(round(result, 3))
elif operator == "*":
    result = num1*num2
    print(round(result, 3))
elif operator == "/":
    result = num1/num2
    print(round(result, 3))
else:
    print(f"Your input '{operator}' is not valid! ")
```

### Python Weight Converter..............

```python
weight = float(input("Enter your weight: "))
unit = input("Kilograms or Pounds? (K or L): ")
if unit == "K":
    weight = weight * 2.205
    unit = "Lbs."
    #print(round(weight, 3))
    print(f"Your weight is: {weight} {unit} ")

elif unit == "L":
    unit = "Kgs."
    weight = weight/2.205
    #print(round(weight, 3))
    print(f"Your weight is: {weight} {unit} ")

else:
    print(f"You entered Wrong {unit} parameter for weight conversion! ")


```
