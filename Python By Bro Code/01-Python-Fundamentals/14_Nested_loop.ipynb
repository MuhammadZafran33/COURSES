### Nested Loop
###### A loop within another loop(inter, outer)


```python
for x in range(1,10):
    print(x, end=" ")
```

```python
for x in range(3):
    for y in range(1,10):
        print(y, end=" ")
    print()
```

```python
rows= int(input("Enter the # of rows: "))
columns= int(input("Enter the # of columns: "))
symbols= input("Enter a symbol to use: ")


for x in range(rows):
    for y in range(columns):
        print(symbols, end="")
    print()
```
